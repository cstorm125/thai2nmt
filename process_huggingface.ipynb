## Process `scb_mt_en_th_2020` for [HuggingFace Datasets](https://github.com/huggingface/datasets/)

```python
import pandas as pd
import glob

official_mappings = {'fake_reviews_yn': 'generated_reviews_yn', 
 'task_master_corpus': 'task_master_1',
 'fake_reviews_hope': 'generated_reviews_translator',
 'thai_websites_corpus': 'thai_websites', 
 'paracrawl_corpus': 'paracrawl', 
 'singapore_corpus': 'nus_sms',
 'mozilla_corpus': 'mozilla_common_voice', 
 'wikipedia_corpus': 'wikipedia', 
 'fake_reviews_crowd': 'generated_reviews_crowd',
 'assorted_government_corpus': 'assorted_government', 
 'aqdf_corpus':'aqdf', 
 'paraphrase_corpus':'msr paraphrase',
 'mfa_corpus': 'assorted_government'}
```

```python
fnames = glob.glob('raw_data/*')
fnames
```

```python
train = pd.read_csv(fnames[0])
train['subdataset'] = train.sentence_id.map(lambda x: x.split(':')[1])
train['sub_dataset'] = train.subdataset.map(lambda x: official_mappings[x])
train = train[['en','th','subdataset']]
train
```

```python
valid = pd.read_csv(fnames[1])
valid['subdataset'] = valid.sentence_id.map(lambda x: x.split(':')[1])
valid['sub_dataset'] = valid.subdataset.map(lambda x: official_mappings[x])
valid = valid[['en','th','subdataset']]
valid
```

```python
test = pd.read_csv(fnames[2])
test['subdataset'] = test.sentence_id.map(lambda x: x.split(':')[1])
test['sub_dataset'] = test.subdataset.map(lambda x: official_mappings[x])
test = test[['en','th','subdataset']]
test
```

```python
#save
train.to_json('data/trainl.json',orient='records',lines=True)
valid.to_json('data/validl.json',orient='records',lines=True)
test.to_json('data/testl.json',orient='records',lines=True)
```
